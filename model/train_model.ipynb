```python
import pandas as pd

# Load ข้อมูล
df_results = pd.read_csv('../dataset/results.csv')
df_stats = pd.read_csv('../dataset/stats.csv')

# STEP 1: Merge ฝั่ง home
df_merged = df_results.merge(
    df_stats,
    how='left',
    left_on=['home_team', 'season'],
    right_on=['team', 'season']
)

# STEP 2: Rename คอลัมน์ซ้ำสำหรับฝั่ง home
important_cols = ['wins', 'losses', 'goals', 'total_yel_card', 'goals_conceded']
df_merged = df_merged.rename(columns={col: col + '_home' for col in important_cols})

# STEP 3: Merge ฝั่ง away ต่อจาก df_merged
df_merged = df_merged.merge(
    df_stats,
    how='left',
    left_on=['away_team', 'season'],
    right_on=['team', 'season'],
    suffixes=('', '_away')
)

# STEP 4: Rename คอลัมน์ซ้ำสำหรับฝั่ง away
df_merged = df_merged.rename(columns={col: col + '_away' for col in important_cols})

# STEP 5: ทดสอบดูผลลัพธ์
df_merged[['home_team', 'away_team', 'wins_home', 'goals_home', 'wins_away', 'goals_away']].head()
```
